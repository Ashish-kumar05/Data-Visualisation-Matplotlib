```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1. Line Graph

# plt.plot(x,y) = x nd y are the axix
# plt.show() = to showing the graph
```

```python
x = np.arange(1,11)
print(x)
```

```python
y = 3*x
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Adding TITLE and LABLES. Changing Line Aesthetics.

plt.plot(x,y, color= "r", linestyle= ":", linewidth= 5)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Adding two lines in same graph

print(x)
```

```python
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1, color="r", linestyle= ":", linewidth= 5)
plt.plot(x,y2, color= "g", linewidth= 3)
plt.title("Two Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Adding sublpot graphs.
# plt.subplot(1,2,1) Function = (1 row, 2 colom, 1 index position)
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

plt.subplot(1,2,1)
plt.plot(x,y1, color= "r", linestyle= ":", linewidth= 5)

plt.subplot(1,2,2)
plt.plot(x,y2, color= "g", linewidth= 3)
plt.title("Sub plot Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, color= "r", linestyle= ":", linewidth= 5)
plt.title("Sub plot Graph")

plt.subplot(2,1,2)
plt.plot(x,y2, color= "g", linewidth= 3)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.bar(x,y)
```

```python
# Bar Graph

Data = {'Ashish':80, 'Arun':67, 'Jay':98, 'Sanjay':89}

Name = list(Data.keys())
Marks = list(Data.values())

plt.bar(Name, Marks, color="g")
plt.title("Bar Graph")
plt.xlabel("Names Of Student")
plt.ylabel("Marks Of Student")
plt.grid(True)
plt.show()
```

```python
plt.barh(Name, Marks, color="r") # Horizontal Bar Graph
plt.title("Bar Graph")
plt.xlabel("Names Of Student")
plt.ylabel("Marks Of Student")
plt.grid(True)
plt.show()
```

```python
# Scatter Graph

x = [6, 3, 2, 5, 8, 1, 7, 10, 9, 4]
y = [17, 12, 22, 11, 21, 27, 19, 28, 14, 23]

plt.scatter(x,y)
plt.show()
```

```python
# Changing Mark Aesthetics.

plt.scatter(x,y, marker="*", color="r", s=100)
plt.title("Scatter Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
# Adding Two Scatter plot.

x = [6, 3, 2, 5, 8, 1, 7, 10, 9, 4]
y = [17, 12, 22, 11, 21, 27, 19, 28, 14, 23]
z = [34, 30, 27, 36, 31, 35, 33, 25, 36, 28]

plt.scatter(x,y, marker="*", color="r", s=100)
plt.scatter(x,z, color="g", s=50)
plt.title("Scatter Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Scatter Sub-Plot

x = [6, 3, 2, 5, 8, 1, 7, 10, 9, 4]
y = [17, 12, 22, 11, 21, 27, 19, 28, 14, 23]
z = [34, 30, 27, 36, 31, 35, 33, 25, 36, 28]

plt.subplot(1,2,1)
plt.scatter(x,y, marker="*", color="r", s=100)
plt.subplot(1,2,2)
plt.scatter(x,z, color="g", s=50)
plt.show()
```

```python
# Histogram
x = (6,6,6,6,3,3,3,3,3,4,4,4,4,2,2,2,2,2,2,2,2,5,8,1,7,7,7,7,10,9,4)

plt.hist(x)
plt.show()
```

```python
# Changing Aesthetics

x = (6,6,6,6,3,3,3,3,3,4,4,4,4,2,2,2,2,2,2,2,2,5,8,1,7,7,7,7,10,9,4)
plt.hist(x, color="orange", bins=4)
plt.show()
```

```python
# BOX PLOT
# plt.boxplot() Function

x = [6, 3, 2, 5, 8, 1, 7, 10, 9, 4]
y = [17, 12, 22, 11, 21, 27, 19, 28, 14, 23]
z = [34, 30, 27, 36, 31, 35, 33, 25, 36, 28]

Data = list([x,y,z])

plt.boxplot(Data)
plt.show()
```

```python
# VIOLIN PLOT
# plt.violinplot() Function = Same like Boxplot

x = [34, 30, 27, 36, 31, 35, 33, 25, 36, 28]
y = [17, 12, 22, 11, 21, 27, 19, 28, 14, 23]
z = [6, 3, 2, 5, 8, 1, 7, 10, 9, 4]

Data1 = list([x,y,z])

plt.violinplot(Data1, showmedians=True)   # If Medial Line required.
plt.title("VIOLIN PLOT")
plt.grid(True)
plt.show()

```

```python
# PIE CHART
# plt.pie(value variable, labels= variable name)

Phone = ['Apple', 'Samsung', 'Oneplus', 'Oppo', 'Vivo']
Sales = [34, 65, 50, 26, 75]
plt.figure(figsize=(6,6))
plt.pie(Sales, labels= Phone)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(Sales, labels= Phone, autopct = "%0.2f%%", colors = ['green', 'red', 'blue', 'orange', 'yellow'])
plt.title("Sales of the Month")
plt.show()
```

```python
# DOUGHNUT - CHART

Phone = ['Apple', 'Samsung', 'Oneplus', 'Oppo', 'Vivo']
Sales = [34, 65, 50, 26, 75]

plt.pie(Sales, labels= Phone, autopct = "%0.2f%%", radius=2.2)
plt.pie([100], colors="w", radius=1)
plt.show()
```
